```python
#Load the raw data
import pandas as pd

df = pd.read_excel(
    "../data/raw/online_retail_II.xlsx",
    sheet_name=None,      # Load ALL sheets at once; returns a dict
    dtype={"Customer ID": str, "StockCode": str}  # Force strings to avoid numeric parsing
)

# Combine both sheets into one DataFrame
df = pd.concat(df.values(), ignore_index=True)
print(f"Total rows: {len(df):,}")
print(f"Columns: {list(df.columns)}")
```
